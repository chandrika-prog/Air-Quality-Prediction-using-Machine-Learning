```python
def cal_SOi(so2):
    si=0
    if(so2<=40):
        si=so2*(50/40)
    elif(so2>40 and so2<=80):
        si=50+(so2-40)*(50/40)
    elif(so2>80 and so2<=300):
        si=100+(so2-80)*(100/300)
    elif(so2>380 and so2<=800):
        si=200+(so2-380)*(100/420)
    elif(so2>800 and so2<=1600):
        si+=300+(so2-800)*(100/800)
    elif(so2>1600):
        si+=400+(so2-1600)*(100/800)
    return si
df['SOi']=df['so2'].apply(cal_SOi)
data=df[['so2','SOi']]
data.head()
    
```

```python
def cal_NOi(no2):
    ni=0
    if(no2<=40):
        ni=no2*(50/40)
    elif(no2>40 and no2<=80):
        ni=50+(no2-40)*(50/40)
    elif(no2>80 and no2<=180):
        ni=100+(no2-80)*(100/100)
    elif(no2>180 and no2<=280):
        ni=200+(no2-180)*(100/100)
    elif(no2>280 and no2<=400):
        ni+=300+(no2-280)*(100/120)
    else:
        ni+=400+(no2-400)*(100/120)
    return ni
df['NOi']=df['no2'].apply(cal_NOi)
data=df[['no2','NOi']]
data.head()
    
```

```python
def cal_RSPMi(rspm):
    ri = 0
    if rspm <= 30:
        ri = rspm * (50 / 30)
    elif rspm > 30 and rspm <= 60:
        ri = 50 + (rspm - 30) * (50 / 30)
    elif rspm > 60 and rspm <= 90:
        ri = 100 + (rspm - 60) * (100 / 30)
    elif rspm > 90 and rspm <= 120:
        ri = 200 + (rspm - 90) * (100 / 30)
    elif rspm > 120 and rspm <= 250:
        ri = 300 + (rspm - 120) * (100 / 130)
    else:
        ri = 400 + (rspm - 250) * (100 / 130)
    return ri

df['RSPMi'] = df['rspm'].apply(cal_RSPMi)
data = df[['rspm', 'RSPMi']]
data.head()
```

```python
def cal_SPMi(spm):
    si = 0
    if spm <= 50:
        si = spm * (50 / 50)
    elif spm > 50 and spm <= 100:
        si = 50 + (spm - 50) * (50 / 50)
    elif spm > 100 and spm <= 250:
        si = 100 + (spm - 100) * (100 / 150)
    elif spm > 250 and spm <= 350:
        si = 200 + (spm - 250) * (100 / 100)
    elif spm > 350 and spm <= 450:
        si = 300 + (spm - 350) * (100 / 100)
    else:
        si = 400 + (spm - 450) * (100 / 100)
    return si

df['SPMi'] = df['spm'].apply(cal_SPMi)
data = df[['spm', 'SPMi']]
data.head()
```

```python
def calc_AQI(row):
    return max(row['NOi'], row['RSPMi'], row['SPMi'], row['SOi'])

df['AQI'] = df.apply(calc_AQI, axis=1)
data = df[['no2', 'NOi', 'rspm', 'RSPMi', 'spm', 'SPMi', 'so2', 'SOi', 'AQI']]
data.head()
```

```python
def AQI_Range(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Satisfactory"
    elif aqi <= 200:
        return "Moderate"
    elif aqi <= 300:
        return "Poor"
    elif aqi <= 400:
        return "Very Poor"
    else:
        return "Severe"

df['AQI_Range'] = df['AQI'] .apply(AQI_Range)
data = df[['AQI', 'AQI_Range']]
data.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, cohen_kappa_score

 
df['date'] = pd.to_datetime(df['date'])

 
df['DayOfYear'] = df['date'].dt.dayofyear
df['Month'] = df['date'].dt.month
df['Weekday'] = df['date'].dt.weekday
 
X2 = df[['SOi', 'NOi', 'RSPMi', 'SPMi', 'DayOfYear', 'Month', 'Weekday']]
Y2 = df['AQI_Range']
 
X2_train2, X2_test2, Y2_train2, Y2_test2 = train_test_split(X2, Y2, test_size=0.33, random_state=70)
 
log_reg = LogisticRegression().fit(X2_train2, Y2_train2)

 
train_preds2 = log_reg.predict(X2_train2)
test_preds2 = log_reg.predict(X2_test2)

print("Model accuracy on train is:", accuracy_score(Y2_train2, train_preds2))
print("Model accuracy on test is:", accuracy_score(Y2_test2, test_preds2))
print("Kappa Score is:", cohen_kappa_score(Y2_test2, test_preds2))
 
sample_input = [[727, 327.55, 78.2, 100, 100, 4, 2]]  # example date: April 9, 2024
print("Prediction:", log_reg.predict(sample_input))
```

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

df['date'] = pd.to_datetime(df['date'])
df['DayOfYear'] = df['date'].dt.dayofyear
df['Month'] = df['date'].dt.month
df['Weekday'] = df['date'].dt.weekday
 
le = LabelEncoder()
df['AQI_Label'] = le.fit_transform(df['AQI_Range'])

X2 = df[['SOi', 'NOi', 'RSPMi', 'SPMi', 'DayOfYear', 'Month', 'Weekday']]
y2 = df['AQI_Label']
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.3, random_state=70)
```

```python
DT2 = DecisionTreeClassifier().fit(X_train2, y_train2)
train_preds3 = DT2.predict(X_train2)
print("Model accuracy on train is: ", accuracy_score(y_train2, train_preds3))
test_preds3 = DT2.predict(X_test2)
print("Model accuracy on test is: ", accuracy_score(y_test2, test_preds3))
print("Kappascore is: ", cohen_kappa_score(y_test2, test_preds3))
```
